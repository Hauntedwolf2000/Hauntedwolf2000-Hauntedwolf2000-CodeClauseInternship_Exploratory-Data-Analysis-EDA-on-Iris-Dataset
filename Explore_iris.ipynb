```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = pd.read_csv("E:/to be removed/CodeClause/iris dataset/iris.csv")

print("Display the first few rows of the dataset")
iris_df.head(5)
```

```python
print("Check the structure of the dataset")
print("\nDataset structure:")
print(iris_df.info())
```

```python
print("Summary statistics")
print("\nSummary statistics:")
print(iris_df.describe())
```

```python
print("Check for missing values")
print("\nMissing values:")
print(iris_df.isnull().sum())
```

```python
print("Visualizing distributions of numerical features")
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(data=iris_df, x="sepal_length", kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(data=iris_df, x="sepal_width", kde=True, color='salmon')
plt.title('Distribution of Sepal Width')
```

```python
plt.subplot(2, 2, 3)
sns.histplot(data=iris_df, x="petal_length", kde=True, color='gold')
plt.title('Distribution of Petal Length')
```

```python
plt.subplot(2, 2, 4)
sns.histplot(data=iris_df, x="petal_width", kde=True, color='greenyellow')
plt.title('Distribution of Petal Width')
```

```python
plt.tight_layout()
plt.show()
```

```python
print("Pairplot to visualize relationships between features")
sns.pairplot(iris_df, hue="species")
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
plt.show()
```

```python
print("Boxplot to visualize distribution of numerical features by species")
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(data=iris_df, x="species", y="sepal_length")
plt.title('Sepal Length by Species')
```

```python
plt.subplot(2, 2, 2)
sns.boxplot(data=iris_df, x="species", y="sepal_width")
plt.title('Sepal Width by Species')
```

```python
plt.subplot(2, 2, 3)
sns.boxplot(data=iris_df, x="species", y="petal_length")
plt.title('Petal Length by Species')
```

```python
plt.subplot(2, 2, 4)
sns.boxplot(data=iris_df, x="species", y="petal_width")
plt.title('Petal Width by Species')
```

```python
print("Exclude non-numeric columns from correlation calculation")
numeric_columns = iris_df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = iris_df[numeric_columns].corr()

print("Plot the correlation heatmap")
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
